# Project 2: Traverse Folder Tree & Filter PDF Files

**Objective**  
Traverse the `/content` folder, find PDF files inside sub-folders `One`, `Two`, and `Three`, extract their text content, and write the extracted text to an `output.txt` file inside each respective sub-folder.

```python
import os
from PyPDF2 import PdfReader
```

```python
main_folder = './content'
sub_folders = ['One', 'Two', 'Three']
```

```python
if not os.path.exists(main_folder):
    print(f"'{main_folder}' not found. Creating it..")
    os.makedirs(main_folder)

for sub in sub_folders:
    sub_path = os.path.join(main_folder, sub)
    if not os.path.exists(sub_path):
        os.makedirs(sub_path)
        print("Created sub-folder: {sub_path}")
    else:
        print(f"Sub_folder '{sub}' already exists.")
```

```python
def read_pdfs_in_folder(folder_path):
    pdf_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.pdf')]

    if not pdf_files:
        print(f"No PDF files found in '{folder_path}'")

    combined_text  = ""
    
    for pdf in pdf_files:
        pdf_path = os.path.join(folder_path, pdf)
        
        try:
            pdf_reader = PdfReader(pdf_path)
            for page in pdf_reader.pages:
                combined_text += page.extract_text() + '\n'
            print(f"Read: '{pdf}'")

        except Exception as e:
            print(f"Exception while reading '{pdf}' : '{e}'")

    return combined_text
```

```python
for sub in sub_folders:
    sub_path = os.path.join(main_folder, sub)
    output_file = os.path.join(sub_path, 'output.txt')

    # Read PDFs in the sub-folder
    text = read_pdfs_in_folder(sub_path)
    
    if not text:
        print(f"Skipping '{sub}' ‚Äî no content to write.\n")
        continue

    # Write extracted text to output.txt
    try:
        f = open(output_file, 'w', encoding='utf-8')
        f.write(text)
        print(f"üìù Written to '{output_file}' successfully!\n")
    except Exception as e:
        print(f"‚ùå Error writing to '{output_file}': {e}\n")
```
